```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from keras import models
from keras import layers
from keras import optimizers
```

# Data Base
- The example trains an **Deep neural Network** to forecast the number of chickenpox cases, taken from the data set **chickenpox_dataset**, given the number of cases in previous months
- Load the example data from **data_chickenpox.txt** text file which contains a single time series, with time steps corresponding to months and values corresponding to the number of cases 
- Construct the input **Deep Neural Network** input w.r.t delay order or output prediction order
- Split the data in **Training Set**, **Validation Set** and **Test Set**
- $nn_0$ gives the output prediction order or number of neurons in the input layer
- $nn_1$ gives the number of neurons in the hidden layer
- $nn_2$ gives the number of neurons in the output layer     

```python
S= np.loadtxt('data_chickenpox.txt')
n=len(S)
plt.plot(S)
plt.title('Chickenpox data')
plt.xlabel('in months')
plt.ylabel('number of cases')
plt.show()

p=1                                      # l'horizon de prdiction
dy=7                                     # la mmoire de sortie / l'ordre de retard
nn_0=dy+1                                # nombre de neurones dans la couche d'entre
nn_1=10                                  # Nombre de neurones dans la couche cache
nn_2=1                                   # Nombre de neurones dans la couche de sortie   
y=S

# Input Signal Contruction

V=y[dy:len(y)-p,]
for i in range(1, dy+1):
    V=np.c_[V,y[dy-i:len(y)-p-i,]]   

T=np.mat(y[dy+p:len(y),]).T                                   # Target Signal Contruction 
print(V.shape)
print(T.shape)
input_NN=V[:400,:]
target_NN=T[:400,]
input_Test=V[400:,:]
target_Test=T[400:,]
```

# Standardize Data
For a better fit and to prevent the training from diverging, standardize the training data to have zero mean and unit variance. Standardize the test data using the same parameters as the training data.

```python
input_mean = np.mean(input_NN)
target_mean = np.mean(target_NN)
input_std = np.std(input_NN)
target_std = np.std(target_NN)

input_NN = (input_NN - input_mean) / input_std
target_NN = (target_NN - input_mean) / target_std
input_Test = (input_Test - input_mean) / input_std
target_Test = (target_Test - target_mean) / target_std
```

# Deep Neural Network Architecture
- Define the type of DNN: In our case we have chosen the **Sequential** DNN
- Define the type of activation function for each layer (**linear,relu,linear** in our case)
- Define the learning/optimization method used (**RMSprop** in our case)
- Define the criteria to minimize and the related metric (**MSE** and **'mae'** in our case)
- Define the **batch size** and the number of **epochs** (**batch_size=4, epochs=50** in our case)
- $nn_0$ gives the prediction order or number of neurons in the input layer
- $nn_1$ gives the number of neurons in the hidden layer
- $nn_2$ gives the number of neurons in the output layer     

```python
def deep_neural_network_architecture(p, dy):
    nn_0=dy+1
    nn_1=10
    nn_2=1
    y=S

    V=y[dy:len(y)-p,]
    for i in range(1, dy+1):
        V=np.c_[V,y[dy-i:len(y)-p-i,]]   

    T=np.mat(y[dy+p:len(y),]).T
    
    input_NN=V[:400,:]
    target_NN=T[:400,]
    input_Test=V[400:,:]
    target_Test=T[400:,]
    
    input_mean = np.mean(input_NN)
    target_mean = np.mean(target_NN)
    input_std = np.std(input_NN)
    target_std = np.std(target_NN)

    input_NN = (input_NN - input_mean) / input_std
    target_NN = (target_NN - input_mean) / target_std
    input_Test = (input_Test - input_mean) / input_std
    target_Test = (target_Test - target_mean) / target_std
    
    model = models.Sequential(
        [
            layers.Dense(nn_0, activation='linear'),
            layers.Dense(nn_1, activation='relu'),
            layers.Dense(nn_2, activation='linear')
        ]
    )

    model.compile(optimizer=optimizers.RMSprop(learning_rate=0.001), loss='mean_squared_error', metrics=['mae'])

    history = model.fit(input_NN, target_NN, epochs=50, batch_size=4, validation_data=(input_Test, target_Test), verbose=0)

    fig, (fig1, fig2) = plt.subplots(1, 2, figsize=(10, 5))

    fig1.plot(history.history['loss'])
    fig1.plot(history.history['val_loss'])
    fig1.set_title('Model loss(p={}, dy={})'.format(p, dy))
    fig1.set_ylabel('Loss')
    fig1.set_xlabel('Epoch')
    fig1.legend(['Train', 'Test'], loc='upper right')

    fig2.plot(history.history['mae'])
    fig2.plot(history.history['val_mae'])
    fig2.set_title('Model MAE(p={}, dy={})'.format(p, dy))
    fig2.set_ylabel('MAE')
    fig2.set_xlabel('Epoch')
    fig2.legend(['Train', 'Test'], loc='upper right')
    
    plt.show()

for p in range(1, 4):
    for dy in range(7, 11):
        deep_neural_network_architecture(p, dy)
```

# Conclusion

En analysant les rsultats obtenus  partir des diffrentes combinaisons de valeurs de `p` et `dy`, nous pouvons tirer les conclusions suivantes :

- L'augmentation de la valeur de `p` rend la tche de prdiction plus difficile, car le modle doit extrapoler sur une priode plus longue en se basant sur les donnes passes. Cela peut entraner une perte de prcision, car les erreurs peuvent s'accumuler au fil du temps. Par consquent, il est important de trouver un quilibre entre l'utilisation d'une priode plus longue pour la prdiction et la capacit du modle  gnraliser correctement les donnes.

- D'autre part, lorsque la valeur de `dy` augmente, cela signifie que le modle prend en compte des dlais plus longs dans les donnes passes pour prdire les valeurs futures. Cela peut potentiellement amliorer la prcision des prdictions en permettant au modle de capturer des motifs  plus long terme. Cependant, cela peut galement rendre le modle plus complexe et plus difficile  entraner, car il doit apprendre des relations temporelles plus complexes entre les donnes. Il est donc important de trouver un compromis entre la complexit du modle et sa capacit  capturer des tendances  long terme.

En rsum, l'ajustement des valeurs de `p` et `dy` dans la fonction `deep_neural_network_architecture(p,dy)` peut avoir un impact significatif sur les performances de prdiction.
